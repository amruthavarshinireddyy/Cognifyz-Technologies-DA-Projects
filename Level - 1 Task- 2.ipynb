CITY ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("restaurant_dataset.csv")
```

CITY WITH HIGHEST NUMBER OF RESTAURANTS IN THE DATASET

```python
restaurant_count=df.groupby('City')['Restaurant ID'].count()
city_with_highest_restaurants=restaurant_count.idxmax()
print("City with the highest number of restaurants :",city_with_highest_restaurants)
```

AVERAGE RATING OF RESTAURANTS IN EACH CITY

```python
average_rating=df.groupby('City')['Aggregate rating'].mean()
```

```python
average_rating.plot(kind='bar',figsize=(26,8))
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Rating of Restaurants in each city')
plt.show()
```

CITY WITH HIGHEST AVERAGE RATING

```python
city_with_highest_average_rating=average_rating.idxmax()
print("City with  highest average rating :",city_with_highest_average_rating)
```
