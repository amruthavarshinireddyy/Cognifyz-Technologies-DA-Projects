TOP CUISINES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("restaurant_dataset.csv")
```

```python
 cuisine_data=df["Cuisines"]
```

```python
cuisine_counts=cuisine_data.value_counts()
```

TOP THREE MOST COMMON CUISINES IN THE DATASET

```python
top_cuisines=cuisine_counts.head(3)
```

```python
total_restaurants=len(df)
percentage=(top_cuisines / total_restaurants)*100
```

```python
plt.bar(percentage.index, percentage.values)
plt.xlabel("Cuisines")
plt.ylabel("Percentage")
plt.title("Top three Cuisines from the dataset")
plt.show()
```

PERCENTAGE OF RESTAURANTS THAT SERVE EACH OF THE TOP CUISINES

```python
plt.pie(top_cuisines, labels=top_cuisines.index, autopct='%1.1f%%')
plt.title("Percentage of top three Cuisines from the dataset")
plt.show()
```
